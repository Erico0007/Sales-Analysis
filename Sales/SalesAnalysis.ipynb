```python
#Data Cleaning and Preprocessing


"""
Cell generated by Data Wrangler.
"""
def clean_data(df):
    # Remove leading and trailing whitespace in columns: 'Date', 'Month' and 12 other columns
    df['Date'] = df['Date'].str.strip()
    df['Month'] = df['Month'].str.strip()
    df['Age_Group'] = df['Age_Group'].str.strip()
    df['Customer_Gender'] = df['Customer_Gender'].str.strip()
    df['Country'] = df['Country'].str.strip()
    df['State'] = df['State'].str.strip()
    df['Product_Category'] = df['Product_Category'].str.strip()
    df['Sub_Category'] = df['Sub_Category'].str.strip()
    df['Product_Description'] = df['Product_Description'].str.strip()
    df[' Unit_Cost '] = df[' Unit_Cost '].str.strip()
    df[' Unit_Price '] = df[' Unit_Price '].str.strip()
    df[' Profit '] = df[' Profit '].str.strip()
    df[' Cost '] = df[' Cost '].str.strip()
    df['Revenue'] = df['Revenue'].str.strip()
    # Replace all instances of "Decmber" with "December" in column: 'Month'
    df['Month'] = df['Month'].str.replace("^Decmber$", "December", regex=True)
    # Replace missing values with 0 in column: 'Day'
    df = df.fillna({'Day': 0})
    # Replace missing values with 0 in column: 'Sales_Order #'
    df = df.fillna({'Sales_Order #': 0})
    # Replace missing values with "N/A" in column: 'Product_Description'
    df = df.fillna({'Product_Description': "N/A"})
    # Replace missing values with "0" in column: 'Age_Group'
    df = df.fillna({'Age_Group': "0"})
    # Replace missing values with 0 in column: 'Order_Quantity'
    df = df.fillna({'Order_Quantity': 0})
    # Replace all instances of "$" with "" in columns: ' Unit_Cost ', ' Unit_Price ' and 2 other columns
    df[' Unit_Cost '] = df[' Unit_Cost '].str.replace("^$$", "", regex=True)
    df[' Unit_Price '] = df[' Unit_Price '].str.replace("^$$", "", regex=True)
    df[' Profit '] = df[' Profit '].str.replace("^$$", "", regex=True)
    df[' Cost '] = df[' Cost '].str.replace("^$$", "", regex=True)
    # Replace all instances of "$" with "" in columns: ' Unit_Cost ', ' Unit_Price ' and 3 other columns
    df.loc[df[' Unit_Cost '].str.lower() == "$".lower(), ' Unit_Cost '] = ""
    df.loc[df[' Unit_Price '].str.lower() == "$".lower(), ' Unit_Price '] = ""
    df.loc[df[' Profit '].str.lower() == "$".lower(), ' Profit '] = ""
    df.loc[df[' Cost '].str.lower() == "$".lower(), ' Cost '] = ""
    df.loc[df['Revenue'].str.lower() == "$".lower(), 'Revenue'] = ""
    # Replace all instances of "$" with "" in column: ' Unit_Cost '
    df[' Unit_Cost '] = df[' Unit_Cost '].str.replace("$", "", case=False, regex=False)
    # Replace all instances of "$" with "" in column: ' Unit_Price '
    df[' Unit_Price '] = df[' Unit_Price '].str.replace("$", "", case=False, regex=False)
    # Replace all instances of "$" with "" in column: ' Profit '
    df[' Profit '] = df[' Profit '].str.replace("$", "", case=False, regex=False)
    # Replace all instances of "$" with "" in column: ' Cost '
    df[' Cost '] = df[' Cost '].str.replace("$", "", case=False, regex=False)
    # Replace all instances of "$" with "" in column: 'Revenue'
    df['Revenue'] = df['Revenue'].str.replace("$", "", case=False, regex=False)
    # Drop duplicate rows across all columns
    df = df.drop_duplicates()
    # Corrected function to categorize age groups
    def transform(col):
        if col < 25:
            return "Youth (<25)"
        elif 25 <= col <= 34:
            return "Young Adult (25-34)"
        elif 35 <= col <= 65:
            return "Adult (35-65)"
        else:
            return "Seniors (65+)"
    # Apply the function to the 'Customer_Age' column
    df['Age_Group'] = df['Customer_Age'].apply(transform)
    # Derive column 'Day,Month,Year' from columns: 'Day', 'Month', 'Year'
    df.insert(5, 'Day,Month,Year', df.apply(lambda row : "", axis=1))
    # Drop columns: 'Day', 'Month', 'Year'
    df = df.drop(columns=['Day', 'Month', 'Year'])
    # Drop column: 'Day,Month,Year'
    df = df.drop(columns=['Day,Month,Year'])
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

```python
#How do zero-cost or zero-price transactions affect profitability?
#convert 'Unit_Cost' and 'Unit_Price' to numeric for analysis
df_clean['Unit_Cost'] = pd.to_numeric(df_clean['Unit_Cost'], errors='coerce')
df_clean['Unit_Price'] = pd.to_numeric(df_clean['Unit_Price'], errors='coerce')

#1. Identify zero-cost transactions
zero_cost_rows = df_clean[(df_clean['Unit_Cost'] == 0)]
#2. Identify zero-price transactions
zero_price_rows = df_clean[(df_clean['Unit_Price'] == 0)]


#3. Calculate Profit for zero-cost transactions
zero_cost_rows['Profit'] = zero_cost_rows['Revenue'] - zero_cost_rows['Cost']
print(f"Average profit for zero-cost transactions: {zero_cost_rows['Profit'].mean()}")


#4. Compare profit of zero-cost transactions to normal rows 
avg_profit_zero_cost = zero_cost_rows['Profit'].mean()
avg_profit_normal = df_clean[(df_clean['Unit_Cost'] > 0) & (df_clean['Unit_Price'] > 0)]['Profit'].mean()


#5 Compare profit of zero-price transactions to normal rows
avg_profit_zero_price = zero_price_rows['Profit'].mean()


#6 Analysis
if avg_profit_zero_cost > avg_profit_normal:
    print("zero-cost rows inflate profitability")
else:
    print("zero-cost rows do not inflate profitability.")


if avg_profit_zero_price < avg_profit_normal:
    print("zero-price rows deflate profitability")
else:
    print("zero-price rows do not deflate profitability.")

#7 Count how many zero-cost and zero-price rows exist
num_zero_cost = len(zero_cost_rows)
num_zero_price = len(zero_price_rows)
print(f"Number of zero-cost rows: {num_zero_cost}")
print(f"Number of zero-price rows: {num_zero_price}")

#8 Evaluate impact on overall dataset
if num_zero_cost > 0 or num_zero_price > 0:
   print(" impact on profitability is significant and should be addressed in analysis.  ")
else:
    print(" impact on profitability is minimal and may not require special handling in analysis.  ")
```

```python
# Are there specific days with unusually high or low sales?

#Filter the dataset for December 2021

data_December = df_clean[(df_clean['Date'] >= '01-12-2021') & (df_clean['Date'] <= '31-12-2021')]

#Group the data

daily_groups = data_December.groupby('Date')

#Calculate total revenue for each day

daily_revenue = daily_groups['Revenue'].sum()

#Compute the Average 

avg_revenue = daily_revenue.mean()

# Compute the standard deviation

std_revenue = daily_revenue.std()

# define the threhold for "Unusual Days"

High_Threshold = avg_revenue + std_revenue
Low_Threshold = avg_revenue - std_revenue

# Identify days with unusually high sales
unusually_high_sales = []
unusually_low_sales = []

#Identify days with unusually high sales

for date in daily_revenue.index:
    if daily_revenue[date] > High_Threshold:
        print(f"{date} had unusually high sales with revenue of {daily_revenue[date]}")
        unusually_high_sales.append(date)
    if daily_revenue[date] < Low_Threshold:
        print(f"{date} had unusually low sales with revenue of {daily_revenue[date]}")
        unusually_low_sales.append(date)



# Output The Results

print(unusually_high_sales)
print(unusually_low_sales)

```
